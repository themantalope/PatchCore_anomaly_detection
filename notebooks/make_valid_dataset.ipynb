```python
import os
import pandas as pd
```

```python
# load up chexpert as the training set
data_path = os.path.join('/','data','matt','chexpert', 'CheXpert-v1.0-small')
os.path.isdir(data_path)
```

```python
file_valid = os.path.join(data_path, 'valid.csv')
os.path.isfile(file_valid)
```

```python
valid_df = pd.read_csv(file_valid)
```

```python
valid_df.head()
```

```python
# make a separate column
valid_df = valid_df.fillna(0)
```

```python
valid_filt = valid_df[valid_df['Frontal/Lateral'] == 'Frontal']
```

```python
normals = []
for i in valid_filt.index:
    no_finding = valid_filt.at[i, 'No Finding']
    device = valid_filt.at[i, 'Support Devices']
    if (int(no_finding) == 1) and (int(device) == 0):
        normals.append(1)
    else:
        normals.append(0)
```

```python
normals
```

```python
valid_filt['Normal'] = normals
```

```python
valid_filt.loc[:, 'Normals'] = normals
```

```python
valid_filt
```

```python
valid_filt['Normal'].mean()
```

```python
outfile = os.path.join(data_path, 'valid_with_normals.csv')
valid_filt.to_csv(outfile, index=False)
```
